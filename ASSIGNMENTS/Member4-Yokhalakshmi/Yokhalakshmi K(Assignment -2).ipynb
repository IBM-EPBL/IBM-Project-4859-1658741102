Find Outliers

```python
qnt = data.drop(columns=['Gender', 'Tenure', 'HasCrCard', 'IsActiveMember', 'NumOfProducts', 'Exited']).quantile(q=[0.25, 0.75])
qnt
```

```python
Q1 = qnt.iloc[0]
Q3 = qnt.iloc[1]
iqr = Q3 - Q1
iqr
```

```python
upper = qnt.iloc[1] + 1.5*iqr
upper
```

```python
lower = qnt.iloc[0] - 1.5*iqr
lower
```

Replace Outliers

```python
data['CreditScore'] = np.where(data['CreditScore']>756, 650.5288, data['CreditScore'])
data['Age'] = np.where(data['Age']>62, 38.9218, data['Age'])
```
